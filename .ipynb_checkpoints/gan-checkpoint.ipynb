```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
from torchvision import transforms
from torchvision import datasets
import torch.optim as opt
from torch.autograd import Variable
```

```python

def to_var(x):
    if torch.cuda.is_available():#utilize the gpu of the pc
        x = x.cuda()
    return Variable(x)

def denorm(x): #It used to restrict a value to a given range.
    out = (x + 1) / 2
    return out.clamp(0, 1)
```

```python
transform=transforms.Compose([
    transforms.ToTensor(),
    #transforms.Normalize((.5,),(.5,))
   
])

mnist=datasets.MNIST(
    root="./NewData/",
    download=True,
    train=True,
    transform=transform 
)
```

```python
dataloaders=torch.utils.data.DataLoader(mnist,shuffle=True,batch_size=64)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
h=128
h2=512
Z=100
lr=1e-3

dataiter=iter(dataloaders)
imgs,labels=dataiter.next()
print(imgs.size())
X=imgs.reshape(imgs.size(0),-1).size(1)
print(X)
```

```python
def imshow(imgs):
    imgs=torchvision.utils.make_grid(imgs)
    npimg=imgs.numpy()
    plt.figure(figsize=(8,8))
    plt.imshow(np.transpose(npimg,(1,2,0)),cmap="Greys_r")
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
imshow(imgs)

    
```

```python
def Xav(m):
    if type(m)==nn.Linear:
         nn.init.xavier_uniform_(m.weight)
         m.bias.data.fill_(0)
    
```

```python
class gen(nn.Module):
    def __init__(self):
        super().__init__()
        self.model=nn.Sequential(
                              nn.Linear(Z,h),
                              nn.ReLU(),
                              nn.Linear(h,X),
                              nn.Sigmoid()
            

        )
        
        self.model.apply(Xav)
        
    def forward(self,input):
        return self.model(input)

```

```python
 G=gen()

```

```python
class dis(nn.Module):
    def __init__(self):
        super().__init__()
        self.model=nn.Sequential(
                                     nn.Linear(X,h),
                     nn.ReLU(),
                     nn.Linear(h,1),
                     nn.Sigmoid()
             
         )
        self.model.apply(Xav)
    
    def forward(self,input):
        return self.model(input)
  
```

```python
D=dis()
print(D)
```

```python
if torch.cuda.is_available():
  D.cuda()
  G.cuda()
```

```python
optimizer_D=opt.Adam(D.parameters(),lr=.0003)
optimizer_G=opt.Adam(G.parameters(),lr=.0003)
criterion = nn.BCELoss()
```

```python

for epoch in range(100):
    G_loss_run=0.
    D_loss_run=0.
    for i,(images,_) in enumerate(dataloaders):
              bs=images.size(0)
              images = to_var(images.view(images.size(0), -1))
              
              z=to_var(torch.randn(bs,Z))
              real_label=to_var(torch.ones(bs,1))
              fake_label=to_var(torch.zeros(bs,1))
              D_real=D(images)
              D_fake=D(G(z))

              D_real_loss=criterion(D_real,real_label)
              D_fake_loss=criterion(D_fake,fake_label)
              D_loss=D_real_loss+D_fake_loss
              optimizer_D.zero_grad()
              D.zero_grad()
              D_loss.backward()
              optimizer_D.step()

              z=to_var(torch.randn(bs,Z))
              D_fake=D(G(z))
              G_loss=criterion(D_fake,real_label)
              optimizer_G.zero_grad()
              D.zero_grad()
              G.zero_grad()
              G_loss.backward()
              optimizer_G.step()

              G_loss_run +=G_loss.item()
              D_loss_run +=D_loss.item()

    print("Epochi={},G_loss={},D_loss={}".format(epoch,G_loss_run/(i+1),D_loss_run/(i+1)))

    sample=G(z).detach() 
    sample=sample.cpu()
    sample=sample.view(sample.size(0),1,28,28)
    imshow(sample)
```

```python

for epochi in range(5):
  for epoch in range(30):
      G_loss_run=0.
      D_loss_run=0.
      for i,(images,_) in enumerate(dataloaders):
              images = to_var(images.view(images.size(0), -1))
              bs=images.size(0)
              z=to_var(torch.randn(bs,Z))
              real_label=to_var(torch.ones(bs,1))
              fake_label=to_var(torch.zeros(bs,1))
              D_real=D(images)
              D_fake=D(G(z))

              D_loss=F.binary_cross_entropy(D_real,real_label)
              D_fake_loss=F.binary_cross_entropy(D_fake,fake_label)
              total=D_loss+D_fake_loss
              optimizer_D.zero_grad()
              total.backward()
              optimizer_D.step()

              z=to_var(torch.randn(bs,Z))
              D_fake=D(G(z))
              G_loss=F.binary_cross_entropy(D_fake,real_label)
              optimizer_G.zero_grad()
              G_loss.backward()
              optimizer_G.step()

              G_loss_run +=G_loss.item()
              D_loss_run +=D_loss.item()

  print("Epochi={},G_loss={},D_loss={}".format(epochi,G_loss_run/(i+1),D_loss_run/(i+1)))

  sample=G(z).detach() 
  sample=sample.cpu()
  sample=sample.view(sample.size(0),1,28,28)
  imshow(sample)
```
